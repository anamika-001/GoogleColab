```
import tensorflow as tf
from tensorflow import keras
```

```
import numpy as np
import pandas as pd
import logging
logger = tf.get_logger()
logger.setLevel(logging.ERROR)
```

```

x=np.random.randint(0,200,(200,1))
#print(x)
y=np.random.randint(0,200,(200,1))
xy=x*y
x2=(x**2)
y2=(y**2)
z=((2*x)+(5*y)+(10*x*y)+15)
p=np.random.randint(0,200,(200,1))
df1=pd.DataFrame(x,columns=['x'])
#df1=pd.DataFrame(x,columns=['y'])
df1['y']=y
df1['xy']=xy
#df1['p']=p
#df1['x^2']=x2
#df1['y^2']=y2
```

```
df1.shape[1]
```

```
model1 = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1,input_shape=[3]),
])
```

```
model1.compile(loss='mean_squared_error',
              optimizer=tf.keras.optimizers.Adam(0.6),
               metrics='MSE')
```

```
from sklearn.model_selection import train_test_split

X_train,X_val,y_train,y_val=train_test_split(df1,z,shuffle=True,test_size=0.33)
```

```
X_train.shape
```

```
history=model1.fit(X_train,y_train,epochs=500, validation_split=0.2,verbose=False)
```

```
val = model1.predict(X_val)
print(val[0:5],y_val[0:5])
```

```
print(history.history)
```

```
X_val[0:5]
```

```

import matplotlib.pyplot as plt
plt.xlabel("Epoch number")
plt.ylabel("Loss Magnitude")
plt.plot(history.history['loss'])
```

```
print("These are the l0 variables: {}".format(model1.get_weights()))
```

```
model1.evaluate(X_train,y_train)
```
